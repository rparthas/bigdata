```python
import pandas as pd
import numpy as np
from pycaret.regression import *
```

```python
train = pd.read_csv('data/HP_train.csv')
test = pd.read_csv('data/HP_test.csv')
# X_train = train.loc[:,train.columns != 'SalePrice']
# y_train = train.loc[:,train.columns == 'SalePrice']
```

```python
reg = setup(data = train, target = 'SalePrice', normalize = True,
            normalize_method='minmax',categorical_imputation='mode',pca=True)
```

```python
model = compare_models(whitelist=['et','rf','catboost','lightgbm','huber'])
```

```python
model=create_model('catboost')
```

```python
model_tuned = tune_model(model)
```

```python
predict_model(model)
```

```python
final_model = finalize_model(model)
pred = predict_model(final_model, data=test)
pred.rename(columns={'Label':'SalePrice'}).loc[:,['Id','SalePrice']].to_csv('data/submission.csv',index=False)
```
