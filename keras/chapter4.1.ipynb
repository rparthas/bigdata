```python
import pandas as pd
feats = pd.read_csv('data/HCV_feats.csv')
target = pd.read_csv('data/HCV_target.csv')
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation
def build_model():
    model = Sequential()
    model.add(Dense(units=4,activation='tanh',input_dim=feats.shape[1]))
    model.add(Dense(units=2,activation='tanh'))
    model.add(Dense(units=1,activation='sigmoid'))
    model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return model
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
classifier = KerasClassifier(build_fn=build_model,epochs =100,batch_size=20,shuffle=False,verbose=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
feats = pd.DataFrame(sc.fit_transform(feats), columns=feats.columns)
```

```python
from sklearn.model_selection import cross_val_score,StratifiedKFold
kf = StratifiedKFold(n_splits=5)
scores = cross_val_score(classifier, feats, target, cv=kf)
print(scores)
```

```python
for x in range(0, 5):
    print(f"Test accuracy at fold {x+1} = {scores[x]:.4f}")
print(f"Final cross validation Test Accuracy : {scores.mean():.4f}")
print(f"Standard Deviation of Final Test Accuracy : {scores.std():.4f}")
```
